```python
#Here missing values are mentioned as '-' so we have to replace it with a variable and then later declare it as a missing value:

#1. If '-' is present in a column then its dtype should be object so at first we have to collect the columns with dtpe object
#Then replace '-' with '0'

cols=[]
for i in df.columns:
  cols.append(f'{i}')
for i in cols:
  if df[i].dtypes=='O':
    df[i]=df[i].str.replace('-','0')
    print(i)
```

```python
#2. Collect the columns which has dtype object only due to the presence of '-'
#Then convert it to float and declare the float 0 as a missing value
#Print the number of missing values for dtype float

cols2=['Contract Until','Club Jersey Number','National Team Jersey Number']

for i in cols2:
  df[f'{i}']=df[f'{i}'].astype(float)
  df.loc[df[f'{i}']==0,f'{i}']=np.NaN
  print(i,'=',df[f'{i}'].isna().sum())
```

```python
#3. Now collect the remaining columns where the values are actually strings
#Then declare the string '-' as a missing value
#Print the number of missing values for dtype object


for i in cols:
  if df[i].dtypes=='O' and i not in cols2:
    df.loc[df[f'{i}']=='0',f'{i}']=np.NaN
    print(i,'=',df[f'{i}'].isna().sum())
```

```python
#Print the number of missing values for dtype int

for i in cols:
  if df[i].dtypes=='int':
    print(i,'=',df[f'{i}'].isna().sum())
```

```python
df.columns
```

```python
#Drop the columns which are not needed for the model along with the columns having large number of missing values

df.drop(['Known As', 'Full Name','Positions Played','Image Link','Club Jersey Number','On Loan','National Team Name','National Team Image Link',
         'National Team Position','National Team Jersey Number'],axis=1,inplace=True)
df.shape
```

```python
#Fill the missing values

df['Contract Until']=df['Contract Until'].fillna(df['Contract Until'].mode()[0])
df['Club Position']=df['Club Position'].fillna(df['Club Position'].mode()[0])
df.isna().sum().unique()
```

```python
#Columns to be encoded

obj_cols=[]
for i in df.columns:
  if df[i].dtypes=='O':
    obj_cols.append(f'{i}')
obj_cols
```

```python
#Get Dummies Encoding

df1=pd.get_dummies(df[['Best Position','Nationality','Club Name','Club Position','Preferred Foot','Attacking Work Rate',
                       'Defensive Work Rate']],dtype=int,drop_first=True)
df1
```

```python
#Concat it with the orginal dataframe

dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
#Drop the columns before encoding

dfe.drop(['Best Position','Nationality','Club Name','Club Position','Preferred Foot','Attacking Work Rate',
          'Defensive Work Rate'],axis=1,inplace=True)
dfe
```

```python
dfe.dtypes.unique()
```

```python
#Seperate x and y

x=dfe.drop(['Value(in Euro)'],axis=1)
x
```

```python
y=dfe['Value(in Euro)']
y
```

```python
#Seperate training and testing data

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
xtrain
```

```python
#Model Creation and evaluation
#Create multiple models to find the best one

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error,r2_score

lr=LinearRegression()
dt=DecisionTreeRegressor()
rf=RandomForestRegressor()

models=[lr,dt,rf]
for i in models:
  i.fit(xtrain,ytrain)
  ypred=i.predict(xtest)
  print('Model started = ',i)
  print('MAPE is',mean_absolute_percentage_error(ytest,ypred))
  print('R2 Score is',r2_score(ytest,ypred))
  print('*'*100)
```

```python
#Here the Random Forest model seems to be the best it
#The R2 Score of 99% indicates it as a good model
#But the MAPE value is very high and it may affect the model's prediction
#It may be due to the presence of Outliers which we have to correct

from scipy import stats

#Compute Z-scores for each column

z_scores = np.abs(stats.zscore(dfe))

#Identify rows with outliers

outliers = (z_scores > 3).any(axis=1)
dfe_outliers = dfe[outliers]

#Number of rows with outliers

print(f"Number of outliers: {dfe_outliers.shape[0]}")
```

```python
#Winsorization - a technique used to limit extreme values in a dataset by clipping data points at the specified percentiles

from scipy.stats.mstats import winsorize

#Create function which limits the winsorization to the 1st and 99th percentiles
#Values below the 1st percentile are set to the value at the 1st percentile
#And values above the 99th percentile are set to the value at the 99th percentile

def winsorize_series(series):
    return winsorize(series, limits=[0.01, 0.01])

#Apply Winsorization to the entire DataFrame

dfe_winsorized = dfe.apply(winsorize_series, axis=0)
dfe_winsorized
```

```python
#Compare original and winsorized data for a specific column

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=dfe['Overall'],color='r')
plt.title('Original Data')

plt.subplot(1, 2, 2)
sns.boxplot(x=dfe_winsorized['Overall'],color='g')
plt.title('Winsorized Data')
plt.show()
```

```python
#Seperate new x and y

x_new=dfe_winsorized.drop(['Value(in Euro)'],axis=1)
x_new
```

```python
y_new=dfe_winsorized['Value(in Euro)']
y_new
```

```python
#Seperate new training and testing data

x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=0.30,random_state=42)
x_train
```

```python
#Create new Random Forest model

rf_new=RandomForestRegressor()
rf_new.fit(x_train,y_train)
y_pred=rf_new.predict(x_test)
y_pred
```

```python
#Evaluation

print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2 Score is',r2_score(y_test,y_pred))
```
